```python
def get_cv_info(index):
    cv_info = {}

    td_elements = all_cv_info[index].find_all("td")
    candidate_link = td_elements[0].find('a')['href']
    cid = candidate_link.split('cid=')[1].split('&')[0]
    td1 = td_elements[1].find('span').get_text()
    name = td_elements[1].find('a').get_text(strip=True)

    td2 = []

    for e in td_elements[1].find_all('span')[1:]:
        text = e.get_text(strip=False)
        td2.append(text)

    td2_text = '\n'.join(td2)


    td3 = []

    for e in td_elements[2].find_all('p'):
        text = e.get_text(strip=False)
        td3.append(text)

    td3_text = '\n'.join(td3)

    td4 = []

    for e in td_elements[3].find_all('li'):
        text = e.get_text(strip=False)
        td4.append(text)

    td4_text = '\n'.join(td4)


    td5 = []

    for e in td_elements[4].find_all('span', class_="text-capitalize"):
        text = e.get_text(strip=False)
        td5.append(text)

    td5_text = '\n'.join(td5)

    cv_info["cid"] = cid
    cv_info["Họ tên"] = name
    cv_info["Trạng thái cv"] = td1
    cv_info["Thông tin"] = td2_text
    cv_info["Kinh nghiệm"] = td3_text
    cv_info["Kỹ năng chuyên môn"] = td4_text
    cv_info["Học vấn"] = td5_text

    return cv_info
```
